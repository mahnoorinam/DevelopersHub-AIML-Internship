# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
Load, inspect, and visualize the Iris dataset to understand data trends and distributions.

## Dataset
Iris Dataset - built into seaborn library (150 flowers, 3 species, 4 features)

```python
# Import all required libraries

import pandas as pd          # for working with data tables
import numpy as np           # for numerical operations
import matplotlib.pyplot as plt  # for creating charts
import seaborn as sns        # for enhanced visualizations

print("Libraries imported successfully!")
```

```python
# Load the Iris dataset directly from seaborn library
# No need to download any file - it's built in

df = sns.load_dataset('iris')

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
df.head()
```

```python
# Check the shape of dataset (rows, columns)
print("Shape of dataset:", df.shape)

# Check column names
print("Column names:", df.columns.tolist())

# Check data types and missing values
print("\nDataset Info:")
df.info()

# Statistical summary (min, max, average etc.)
print("\nStatistical Summary:")
df.describe()
```

```python
# Scatter Plot: Sepal Length vs Sepal Width
# Each dot represents one flower, colored by species

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species',    # different color for each species
    palette='Set1'
)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Histogram: Distribution of all 4 features
# Shows how values are spread across the dataset

df.drop('species', axis=1).hist(
    bins=20,
    figsize=(10, 6),
    color='steelblue',
    edgecolor='black'
)
plt.suptitle('Distribution of All Features', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Box Plot: Shows outliers and data spread per species
# Each box shows min, max, median and outliers for each feature

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    ax = axes[i//2, i%2]
    sns.boxplot(data=df, x='species', y=feature, palette='pastel', ax=ax)
    ax.set_title(f'Box Plot of {feature}')

plt.tight_layout()
plt.show()
```

## Summary of Findings

- The dataset contains 150 rows and 5 columns with no missing values.
- There are 3 species: Setosa, Versicolor, and Virginica — 50 flowers each.
- The scatter plot shows that Setosa is clearly separable from the other two species.
- Histograms reveal that petal_length and petal_width have a bimodal distribution — Setosa has much smaller petals.
- Box plots confirm that Setosa has the smallest sepal length but the widest sepal width among all species.
- A few outliers were detected in sepal_width for Setosa and Virginica species.
